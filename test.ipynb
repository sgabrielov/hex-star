```python
from queue import PriorityQueue
```

```python
def bi_join_nodes(child_f, child_b):
    #print("stuck on bijoin")
    current_node = child_b.parent
    next_node_b = current_node.parent
    parent_node = child_f
    current_node.parent = parent_node
    if next_node_b is None:
        return current_node
    #print("parent_node", parent_node.state[0])
    #print("current_node", current_node.state[0])
    
    while next_node_b is not None:
        #print("current_node", current_node.state[0])
        parent_node = current_node
        current_node = next_node_b
        next_node_b = next_node_b.parent
        current_node.parent = parent_node
    #print("current_node", current_node.state[0])
    return current_node
```

```python
def f_bi_join_nodes(direction, child_f, child_b):
    if direction == "f":
        return bi_join_nodes(child_f, child_b)
    else:
        return bi_join_nodes(child_b, child_f)
```

```python
class Node:
    def __init__(self, state, parent):
        self.state = state
        self.parent = parent

    def __str__(self):
        return str(self.state)
```

```python
n1 = Node(1, None)
n2 = Node(2, n1)
n3 = Node(3, n2)

n4 = Node('a', None)
n5 = Node('b', n4)
n6 = Node('c', n5)

soln = f_bi_join_nodes('d', n3, n6)

tmp = soln
while tmp:
    print(tmp)
    tmp = tmp.parent

```
